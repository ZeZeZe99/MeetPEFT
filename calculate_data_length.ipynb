```python
import tiktoken
import json
import matplotlib.pyplot as plt
```

```python
encoding = tiktoken.get_encoding("cl100k_base") # gpt-3.5-turbo
```

```python
def num_tokens_from_string(string):
    return len(encoding.encode(string))
```

## MeetingBank

```python
path_mb = "MeetingBank/"
files = ["test", "train", "validation"]
```

```python
for file in files:
    length_list = []
    data_path = path_mb + file + ".json"
    with open(data_path, "r") as f:
        data = json.load(f)
        for item in data:
            length_list.append(num_tokens_from_string(item["source"]))
    plt.hist(length_list, bins=100)
    plt.xlabel("Number of tokens")
    plt.ylabel("Times")
    plt.title(path_mb + file)
    plt.savefig(path_mb[:-1] + "_" + file + ".png")
    plt.show()
```

## QMSum

```python
path_qm = "QMSum-main/MEETPEFT_data/ALL/jsonl/"
files = ["test", "train", "val"]
```

```python
for file in files:
    length_list = []
    data_path = path_qm + file + ".jsonl"
    with open(data_path, "r") as f:
        for line in f:
            item = json.loads(line)
            length_list.append(num_tokens_from_string(item["conversations"]))
    plt.hist(length_list, bins=100)
    plt.xlabel("Number of tokens")
    plt.ylabel("Times")
    plt.title(path_mb + file)
    plt.savefig(path_mb[:-1] + "_" + file + ".png")
    plt.show()
```
